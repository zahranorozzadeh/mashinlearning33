```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
N=1000
std = 0.6
banana_width = np.random.normal(4,std,N)
banana_length = np.random.normal(8,std,N)


apple_width = np.random.normal(6,std,N)
apple_length = np.random.normal(6,std+0, N)


width = np.concatenate((banana_width,apple_width))
length = np.concatenate((banana_length,apple_length))


X = np.array([width,length]).T

banana_label = np.zeros(N, dtype='int')
apple_label = np.ones(N, dtype='int')

Y= np.concatenate((banana_label,apple_label))
plt.scatter(X[0:N, 1], X[0:N, 0], c='orange', label='banana')
plt.scatter(X[N:2*N, 1], X[N:2*N, 0], c='red', label='apple')


plt.xlabel('length')
plt.ylabel('width')
plt.title("Apple and Banana")
plt.legend()
```

```python
labels = ['banana', 'apple', 'banana', 'apple']
x = [1, 1.5, 2.5, 3.0]
height = [banana_width.mean(),apple_width.mean(),banana_length.mean(),apple_length.mean()]
yerr = [banana_width.std(),apple_width.std(),banana_length.std(),apple_length.std()]

color = ['tab:blue', 'tab:blue', 'tab:orange', 'tab:orange', 'tab:red', 'tab:red']

plt.bar(x=x,
        height=height,
        color=color,
        width=0.35,
        error_kw={'lw': 4},
        yerr=yerr)
plt.xticks(ticks=x, labels=labels)
plt.title('Fruit length VS width VS Weight')

plt.grid()
```

```python
fea = 'length'
plt.boxplot(x=banana_length,patch_artist = True,boxprops={'facecolor':'tab:blue'},positions=[0],labels=['bananas'],widths=[0.15])
plt.boxplot(x=apple_length,patch_artist = True,boxprops={'facecolor':'tab:red'},positions=[0.3],labels=['apple'],widths=[0.15])

plt.ylabel('Length')
plt.xlim(-0.2,0.5)
plt.title('length Distribution')
```

```python
fea = 'width'
plt.hist(x=banana_length,
         bins=30, density=True, color='tab:blue', label='banana',)
plt.hist(x=apple_length,
         bins=30, density=True, color='tab:orange', label='apple', alpha=0.8)
plt.legend(frameon=False)
plt.title('Width Distribution ')
plt.xlabel('Width')
```

```python

fig, (ax, cax) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [0.9, 0.1]})
_, _, _, mappable = ax.hist2d(
    x=length, y=width, bins=50, cmap=plt.cm.hot)
ax.set_xlabel('length')
ax.set_ylabel('width')

plt.colorbar(mappable=mappable, cax=cax, ax=ax) 
```

```python
fea = 'L/W'
fea1 = banana_length/banana_width
fea2 = apple_length/apple_width

plt.hist(x=fea1,
       bins=30, density=True, color='tab:blue', label='banana' )


plt.hist(x=fea2,
       bins=30, density=True, color='tab:orange', label='apple' )

plt.legend()
plt.title('L/W Distribution')
plt.xlabel('L/W ratio')
```

```python
fea = 'area'
fea1 = banana_length*banana_width
fea2 = apple_length*apple_width

plt.hist(x=fea1, histtype='step', lw=4,
         bins=30, density=True, color='tab:blue', label='bananas')
plt.hist(x=fea2, histtype='step', lw=4,
         bins=30, density=True, color='tab:orange', label='apples')
plt.title('Area Distribution ')
plt.legend()
plt.xlabel('Area')
```

```python
fea = 'circumference'
fea1 =2*(banana_length+banana_width) 
fea2 = 2*(apple_length+apple_width)

plt.hist(x=fea1, histtype='step', lw=4,
         bins=30, density=True, color='tab:blue', label='bananas')
plt.hist(x=fea2, histtype='step', lw=4,
         bins=30, density=True, color='tab:orange', label='apples')
plt.title('circumference Distribution ')
plt.legend()
plt.xlabel('circumference')
```
