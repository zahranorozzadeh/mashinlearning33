```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 100

melon_width = np.random.normal(40, 15, n)
melon_length = np.random.normal(20, 10, n)
melon_weight = np.random.normal(8000, 300, n)

balloons_width = np.random.normal(100, 5, n)
balloons_length = np.random.normal(100, 5, n)
balloons_weight = np.random.normal(100, 10, n)

width = np.concatenate((melon_width, balloons_width))
length = np.concatenate((melon_length, balloons_length))
weight = np.concatenate((melon_weight, balloons_weight))

X = np.array([width, length, weight]).T

balloons_label = np.zeros(n, dtype='int')
melons_label = np.ones(n, dtype='int')

Y = np.concatenate((balloons_label, melons_label))
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X[:n,0], X[:n,1], X[:n,2], c= 'yellow')
ax.scatter(X[n:,0], X[n:,1], X[n:,2], c= 'red')

ax.set_xlabel('width')
ax.set_ylabel('length')
ax.set_zlabel('weight')
```
